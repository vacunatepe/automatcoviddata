Cargando Contagios





```
# Cargando algunas librerias
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
# Cargando los datos
url1 = "https://raw.githubusercontent.com/vacunatepe/automatcoviddata/main/contagiosUS.csv"
cont = pd.read_csv(url1,dtype={'UBIGEO': str})
cont = cont.drop(['Unnamed: 0'], axis=1)
cont = cont.rename(columns={'UUID': 'CONTAGIOS'})
cont
```

Cargando Vacunas

```
url2 = "https://raw.githubusercontent.com/vacunatepe/automatcoviddata/main/vacunaUS.csv" 
vac = pd.read_csv(url2,dtype={'UBIGEO': str})
vac = vac.drop(['Unnamed: 0'], axis=1)
vac = vac.rename(columns={'UUID': 'VACUNADOS'})
vac
```

Uniendo vacunas con contagio

```
new_df = pd.merge(cont, vac,  how='left', left_on=['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'], right_on = ['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'])
new_df['VACUNADOS'] = new_df['VACUNADOS'].fillna(0)
new_df
```

Cargando movilidad

```
url3 = "https://raw.githubusercontent.com/vacunatepe/automatcoviddata/main/movilidad.csv" 
mov = pd.read_csv(url3,dtype={'UBIGEO': str})
mov = mov.drop(['Unnamed: 0'], axis=1)
mov = mov.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'CENTRO_COMERCIAL',
                          'grocery_and_pharmacy_percent_change_from_baseline':'TIENDAS_FARMACIAS',
                          'parks_percent_change_from_baseline':'PARQUES',
                          'transit_stations_percent_change_from_baseline':'PARADEROS',
                          'workplaces_percent_change_from_baseline':'CENTRO_LABORAL'})
mov = mov[['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA','CENTRO_COMERCIAL','TIENDAS_FARMACIAS','PARQUES','PARADEROS','CENTRO_LABORAL']]
mov
```

Combinando contagios, vacunas y movilidad

```
new_df2 = pd.merge(new_df, mov,  how='left', left_on=['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'], right_on = ['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'])
new_df2
```

Cargando fallecidos

```
url4 = "https://raw.githubusercontent.com/vacunatepe/automatcoviddata/main/fallecidosUS.csv" 
fall = pd.read_csv(url4,dtype={'UBIGEO': str})
fall = fall.drop(['Unnamed: 0'], axis=1)
fall = fall.rename(columns={'Fecha': 'FALLECIDOS'})
fall
```

Combinandos Contagio, vacuna, movilidad y fallecidos

```
new_df3 = pd.merge(new_df2, fall,  how='left', left_on=['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'], right_on = ['DEPARTAMENTO', 'PROVINCIA','DISTRITO','UBIGEO','SEMANA'])
new_df3['FALLECIDOS'] = new_df3['FALLECIDOS'].fillna(0)
new_df3
```

Guardando Base

```
new_df3.to_csv('datossemana.csv')
```
